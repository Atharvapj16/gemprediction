```python
#Define the custom ranking for each ordinal variable
cut_cat=['Fair','Good','Very Good','Premium','Ideal']
color_cat=['D','E','F','G','H','I','J']
clarity_cat=['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF']
```

```python
from sklearn.impute import SimpleImputer         #######  For handling missing values
from sklearn.preprocessing import StandardScaler #######  For handling feature scaling (i.e. to bring data in specific range)
from sklearn.preprocessing import OrdinalEncoder #######  For ordinal encoding

###Piplines

from sklearn.pipeline  import  Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline(

    steps=[
        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('scaler',StandardScaler())
    ]
)
```

```python
cat_pipeline=Pipeline( 

    steps=[

        ('imputer',SimpleImputer(strategy='most_frequent')),
        ('scaler',OrdinalEncoder(categories=[cut_cat,color_cat,clarity_cat]))
        ]
)
```

```python
preprocesor=ColumnTransformer(
    [
        ('num_pipeline',num_pipeline,numerical_cols),
        ('cat_pipeline',cat_pipeline,categorical_cols)
    ]
)
```

```python
#Train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=30)
```

```python
preprocesor.fit_transform(X_train)
```

```python
preprocesor.transform(X_test)
```

```python
preprocesor.get_feature_names_out()
```

```python
X_train=pd.DataFrame(preprocesor.fit_transform(X_train),columns=preprocesor.get_feature_names_out())
X_test=pd.DataFrame(preprocesor.transform(X_test),columns=preprocesor.get_feature_names_out())
```

```python
### Applying Lasso,Ridge,Elastic net regression
```

```python
from  sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import numpy as np
def evaluate_model(true,predicted):
    mae=mean_absolute_error(true,predicted)
    mse=mean_squared_error(true,predicted)
    rsme=np.sqrt(mean_squared_error(true,predicted))
    r2_square=r2_score(true,predicted)
    return mae,rsme,r2_square
```

```python
models={
       'LinearRegression':LinearRegression(),
       'Lasso':Lasso(),
       'Ridge':Ridge(),
       'ElasticNet':ElasticNet()
       }
```

```python
trained_model_list=[]
model_list=[]
r2_list=[]
```

```python
list(models)
```

```python
for i in range(len(list(models))):
    model=list(models.values())[i]
    print(models)
```

```python

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,Y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    #this is a validation(test) score
    mae, rmse, r2_square=evaluate_model(Y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
